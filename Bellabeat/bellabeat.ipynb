## Data Pre-proccesing and Exploring Data

```python
bela = bela[['Id','ActivityDate','TotalSteps','VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes','Calories']]
bela
```

```python
bela.rename(columns={'ActivityDate':'Date',},inplace=True)
bela
```

```python
bela['TotalMinutes']=bela['VeryActiveMinutes']+bela['FairlyActiveMinutes']+bela['LightlyActiveMinutes']+bela['SedentaryMinutes']
bela
```

```python
bela['TotalHours'] = round(bela['TotalMinutes']/60,0)
bela
```

```python
bela.info()
```

```python
bela['Date'] =pd.to_datetime(bela['Date'])
```

```python
bela.info()
```

```python
bela['Day'] = bela['Date'].dt.year
bela
```
